```python
import numpy as np
from typing import Optional
import matplotlib.pyplot as plt
```

```python
class Die:
    def __init__(self, n_sides: int, seed: Optional[int] = None):
        self.rng = np.random.default_rng(seed=seed)
        self.n_sides = n_sides
    
    def __call__(self) -> int:
        return int(self.rng.uniform(1, self.n_sides + 1))

    def roll_with_bonus(self, bonus: int) -> int:
        return self() + bonus
    
    def roll_with_advantage(self, bonus: int) -> int:
        return max(self(), self()) + bonus
```

```python
d6 = Die(6)
```

```python
print(d6.roll_with_advantage(0))
```

```python
plt.hist([d6() for _ in range(int(1e6))], bins=[1, 2, 3, 4, 5, 6, 7])
plt.xticks([1.5, 2.5, 3.5, 4.5, 5.5, 6.5], ["1", "2", "3", "4", "5", "6"])
```

```python
plt.hist([d6.roll_with_advantage(0) for _ in range(int(1e6))], bins=[1, 2, 3, 4, 5, 6, 7])
plt.xticks([1.5, 2.5, 3.5, 4.5, 5.5, 6.5], ["1", "2", "3", "4", "5", "6"])
```
